```python
import numpy 

from matplotlib import pyplot                 
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
def linearconv(nt):
 
    Vmax = 80    # km/hour - maximum traffic velocity
    L = 11       # km - length of highway lane
    rhomax = 250 # cars/km - maximum car density per length of highway
    nx = 51      # distance step difference
    dt = .001    # hours - time step difference

    dx = L/(nx-1)
    x = numpy.linspace(0,L,nx)
    rho = numpy.ones(nx)*10
    rho[10:20] = 50
    rho[0] = 10 #intial boundary condition
    
    V = Vmax*(1-rho/rhomax)*1000/3600
        
    for n in range(nt):  
        rhon = rho.copy() 
        rho[1:] = rhon[1:] -Vmax*dt/dx*(rhon[1:]*(1-rhon[1:]/rhomax) -rhon[0:-1]*(1-rhon[0:-1]/rhomax)) 
        rho[0] = 10
        V = Vmax*(1-rho/rhomax)*1000/3600
    
    pyplot.figure(figsize=(6,6))   #set plot size
    pyplot.plot(x, rho, color='#003366', ls='--', lw=3)
    pyplot.ylim(0,numpy.amax(rho)+1);
    pyplot.xlim(0,L);
    #pyplot.tick_params(axis='both', labelsize=14) #increase font size for ticks
    pyplot.xlabel('Distance along Highway, $x$ in meters', fontsize=14) #x label
    pyplot.ylabel('Car Density, $rho$ in cars/kilometer', fontsize=14) #y label

    pyplot.figure(figsize=(6,6))   #set plot size
    pyplot.plot(x, V, color='#003366', ls='--', lw=3)
    pyplot.ylim(0,numpy.amax(V)+1);
    pyplot.xlim(0,L);
    #pyplot.tick_params(axis='both', labelsize=14) #increase font size for ticks
    pyplot.xlabel('Distance along Highway, $x$ in meters', fontsize=14) #x label
    pyplot.ylabel('Speed of Traffic, $V$ in kilometers/hour', fontsize=14) #y label
    
    print("Plot for Minute",nt*dt*60)
    print("Minimum Speed of Traffic is", round(numpy.amin(V),2), "meters/second")
    print("Average Speed  of Traffic is", round(numpy.average(V),2), "meters/second")
```

Enter the *minimum* velocity at time t = 0 in meters per second. Round to 2 decimal places

```python
linearconv(0)
```
